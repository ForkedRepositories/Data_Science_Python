```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```python
print(cv2.__version__)

face_cascade_path =  r"C:\Users\caihao.cui\Documents\Python Scripts\opencv_py\opencv-master\data\haarcascades\haarcascade_frontalface_default.xml"

print(face_cascade_path)

face_cascade = cv2.CascadeClassifier(face_cascade_path)
```

```python
eye_cascade_path =  r"C:\Users\caihao.cui\Documents\Python Scripts\opencv_py\opencv-master\data\haarcascades\haarcascade_eye.xml"
print(eye_cascade_path)
eye_cascade = cv2.CascadeClassifier(eye_cascade_path)
```

```python
img=mpimg.imread('MyLTUTeam.jpg')
plt.imshow(img)
```

```python
img = cv2.imread('MyLTUTeam.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# OpenCV represents RGB images as multi-dimensional NumPy arrays…but in reverse order!
# This means that images are actually represented in BGR order rather than RGB!
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, 'gray', )
```

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        
plt.figure()        
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
